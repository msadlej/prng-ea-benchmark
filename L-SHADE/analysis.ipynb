## Wilcoxon signed-rank test

```python
lcg_errors = combine_errors(lcg)
mt_errors = combine_errors(mt)
true_errors = combine_errors(true)

print("True Random vs Linear Congruential:")
display(calculate_wilcoxon(true_errors, lcg_errors, styled=True))

print("\nTrue Random vs Mersenne Twister:")
display(calculate_wilcoxon(true_errors, mt_errors, styled=True))
```

### Linear congruential

```python
lcg_1M_errors = combine_errors(lcg_1M)
lcg_100K_errors = combine_errors(lcg_100K)
lcg_10K_errors = combine_errors(lcg_10K)
lcg_1K_errors = combine_errors(lcg_1K)
lcg_100_errors = combine_errors(lcg_100)

print("2.5M vs 1M Linear Congruential:")
display(calculate_wilcoxon(lcg_errors, lcg_1M_errors, styled=True))

print("\n2.5M vs 100K Linear Congruential:")
display(calculate_wilcoxon(lcg_errors, lcg_100K_errors, styled=True))

print("\n2.5M vs 10K Linear Congruential:")
display(calculate_wilcoxon(lcg_errors, lcg_10K_errors, styled=True))

print("\n2.5M vs 1K Linear Congruential:")
display(calculate_wilcoxon(lcg_errors, lcg_1K_errors, styled=True))

print("\n2.5M vs 100 Linear Congruential:")
display(calculate_wilcoxon(lcg_errors, lcg_100_errors, styled=True))
```

### Mersenne Twister

```python
mt_1M_errors = combine_errors(mt_1M)
mt_100K_errors = combine_errors(mt_100K)
mt_10K_errors = combine_errors(mt_10K)
mt_1K_errors = combine_errors(mt_1K)
mt_100_errors = combine_errors(mt_100)

print("2.5M vs 1M Mersenne Twister:")
display(calculate_wilcoxon(mt_errors, mt_1M_errors, styled=True))

print("\n2.5M vs 100K Mersenne Twister:")
display(calculate_wilcoxon(mt_errors, mt_100K_errors, styled=True))

print("\n2.5M vs 10K Mersenne Twister:")
display(calculate_wilcoxon(mt_errors, mt_10K_errors, styled=True))

print("\n2.5M vs 1K Mersenne Twister:")
display(calculate_wilcoxon(mt_errors, mt_1K_errors, styled=True))

print("\n2.5M vs 100 Mersenne Twister:")
display(calculate_wilcoxon(mt_errors, mt_100_errors, styled=True))
```

### True Random

```python
true_1M_errors = combine_errors(true_1M)
true_100K_errors = combine_errors(true_100K)
true_10K_errors = combine_errors(true_10K)
true_1K_errors = combine_errors(true_1K)
true_100_errors = combine_errors(true_100)

print("2.5M vs 1M True Random:")
display(calculate_wilcoxon(true_errors, true_1M_errors, styled=True))

print("\n2.5M vs 100K True Random:")
display(calculate_wilcoxon(true_errors, true_100K_errors, styled=True))

print("\n2.5M vs 10K True Random:")
display(calculate_wilcoxon(true_errors, true_10K_errors, styled=True))

print("\n2.5M vs 1K True Random:")
display(calculate_wilcoxon(true_errors, true_1K_errors, styled=True))

print("\n2.5M vs 100 True Random:")
display(calculate_wilcoxon(true_errors, true_100_errors, styled=True))
```
